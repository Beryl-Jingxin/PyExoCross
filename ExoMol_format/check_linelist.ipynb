````python
read_path = '../data/results/exoweb/CN/12C-14N__Trihybrid.states.bz2'

s_df_1 = pd.read_csv(read_path, compression='bz2', sep='\s+', header=None)

s_df_1
````

````python
s_df_1[0] = s_df_1[0].map('{: >12d}'.format)
s_df_1[1] = s_df_1[1].map('{: >12.6f}'.format)
s_df_1[2] = s_df_1[2].map('{: >6d}'.format)
s_df_1[3] = s_df_1[3].map('{: >7.1f}'.format)
s_df_1[4] = s_df_1[4].map('{: >12.6f}'.format)
s_df_1[5] = s_df_1[5].map('{: >12.4E}'.format)
s_df_1[6] = s_df_1[6].map('{: >12.6f}'.format)
s_df_1[7] = s_df_1[7].map('{: >2}'.format)
s_df_1[8] = s_df_1[8].map('{: >2}'.format)
s_df_1[9] = s_df_1[9].map('{: >2}'.format)
s_df_1[10] = s_df_1[10].map('{: >3d}'.format)
s_df_1[11] = s_df_1[11].map('{: >3d}'.format)
s_df_1[12] = s_df_1[12].map('{: >5.1f}'.format)
s_df_1[13] = s_df_1[13].map('{: >5.1f}'.format)
s_df_1[14] = s_df_1[14].map('{: >3}'.format)
s_df_1[15] = s_df_1[15].map('{: >12.6f}'.format)
s_df_1 = s_df_1.sort_values(by=0) 

s_df_1
````

````python
txt_path = '../data/results/exoweb/CN/12C-14N__Trihybrid.txt'
bz2_path = '../data/results/exoweb/CN/12C-14N__Trihybrid.states.bz2'

s_df_1.to_csv(txt_path, sep=',', quoting=csv.QUOTE_NONE, escapechar=',', 
              header=None, index=False, chunksize = 1000)

def read_txt_in_chunks(path, chunk_size=1024*1024):
    file = open(path, 'r')
    while True:
        chunk_data = file.read(chunk_size)
        if not chunk_data:
            break
        yield chunk_data

with bz2.BZ2File(bz2_path, 'w') as bz2_file:
    for chunk in read_txt_in_chunks(txt_path):
        string = str(chunk).replace(',',' ')
        data = bytes(string,'utf-8')
        bz2_file.write(data)
    bz2_file.close()
print('Finished!')
````

### Fill all empty fields with NaN and replace nan to NaN 

````python
read_path = '/mnt/data/exomol/exomol3_data/SO2/32S-16O2/ExoAmes/32S-16O2__ExoAmes.states.bz2'

s_df_1 = pd.read_csv(read_path, compression='bz2', sep='\s+', names=['a','b','c','d','e','f','g','h','i','j','k'],header=None)

s_df_1
````

````python
s_df_2 = pd.DataFrame()
s_df_2[0] = s_df_1['a'].map('{: >12d}'.format)
s_df_2[1] = s_df_1['b'].map('{: >12.6f}'.format)
s_df_2[2] = s_df_1['c'].map('{: >6d}'.format)
s_df_2[3] = s_df_1['d'].map('{: >7.1f}'.format)
s_df_2[4] = s_df_1['e'].map('{: >12.6f}'.format).replace('         nan','         NaN')
s_df_2[5] = s_df_1['f'].map('{: >3d}'.format)
s_df_2[6] = s_df_1['g'].dropna().astype(int).map('{: >3d}'.format)
s_df_2[7] = s_df_1['h'].dropna().astype(int).map('{: >3d}'.format)
s_df_2[8] = s_df_1['i'].dropna().astype(int).map('{: >3d}'.format)
s_df_2[9] = s_df_1['j'].dropna().astype(int).map('{: >3d}'.format)
s_df_2[10] = s_df_1['k'].dropna().astype(int).map('{: >3d}'.format)
s_df_2 = s_df_2.sort_values(by=0) 

s_df_2
````

````python
import numpy as np
s_df_3 = s_df_2.fillna('NaN')
s_df_3
````

````python
txt_path_1 = '/mnt/data/exomol/exomol3_data/SO2/32S-16O2/ExoAmes/32S-16O2__ExoAmes_1.states'
txt_path_2 = '/mnt/data/exomol/exomol3_data/SO2/32S-16O2/ExoAmes/32S-16O2__ExoAmes_2.states'

s_df_3.to_csv(txt_path_1, sep=',', quoting=csv.QUOTE_NONE, escapechar=',', 
              header=None, index=False, chunksize = 1000)

def read_txt_in_chunks(path, chunk_size=1024*1024):
    file = open(path, 'r')
    while True:
        chunk_data = file.read(chunk_size)
        if not chunk_data:
            break
        yield chunk_data

with open(txt_path_2, 'w') as txt_file:
    for chunk in read_txt_in_chunks(txt_path_1):
        string = str(chunk).replace(',',' ')        
        txt_file.write(string)
    txt_file.close()
print('Finished!')
````
